<font face="vazir" size=3>
پس از آماده‌سازی داده‌ها نوبت به ساخت شبکه‌ی عصبی می‌رسد. باز هم دست‌تان باز است تا معماری موردنظر خود را طراحی کنید و از هر تکنیکی که لازم می‌دانید جهت بهتر کردن فرآیند آموزش مدل استفاده کنید. پیشنهاد می‌کنیم تکنیک‌های مختلفی که تا اینجای دوره آموخته‌اید را آزمایش کرده و به تفسیر نتایج بپردازید.

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train.drop(columns = ['satisfaction']), 
                                                             train['satisfaction'],
                                                            test_size = 0.2,
                                                            random_state = 42)
```

```python
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout, Input, Activation

model = Sequential()
model.add(Input(shape=(27,)))
model.add(Dense(80))
#model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(40))
model.add(BatchNormalization())
model.add(Activation('relu'))



model.add(Dense(10))
model.add(BatchNormal ization())
model.add(Activation('relu'))

model.add(Dense(1, activation = 'sigmoid'))     
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
early_stop = EarlyStopping(
    monitor = 'val_accuracy',
    patience=10,
    restore_best_weights=True)
```

```python
history = model.fit(x = X_train, y = y_train, validation_data = (X_val, y_val), epochs  = 30, batch_size = 32,
                   callbacks=[early_stop])
```
